```python
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
%env DATA_FILE=../data/NYPD_Complaint_Data_Historic.csv
```

```python
!xsv headers $DATA_FILE
```

```python
%env STATS_FILE=../data/nypd_data_stats.csv
# create an index on data file
!xsv index $DATA_FILE
# get statistics on everything and pipe them to a stats file
!xsv stats $DATA_FILE --everything > $STATS_FILE
# set a stats file 
!xsv index $STATS_FILE
stats_file = %env STATS_FILE
```

```python
pd.read_csv(stats_file)
```

```python
!xsv count $DATA_FILE
```

```python
%env FREQ_FILE=../data/nypd_data_freq.csv
!xsv frequency $DATA_FILE --limit 10 > $FREQ_FILE
!xsv index $FREQ_FILE
freq_file = %env FREQ_FILE
```

```python
pd.read_csv(freq_file)
```
